```python
from selenium import webdriver
import time
import re

# 크롬 웹브라우저 실행
path = "C:/Users/chlgo/hac/chromedriver.exe"
urls=[]
driver = webdriver.Chrome(path)
#url 은 네이버에 파이썬이라고 검색하고 나온 검색결과 URL 주소로 설정
for i in range(1, 16):  # 1~15페이지의 블로그 글 url 받아오기 위함
    url = 'https://blog.naver.com/PostList.naver?from=postList&blogId=insoonlife&categoryNo=47&currentPage='+str(i) 
    driver.get(url)
    time.sleep(0.5)
    print(url)
    for j in range(1,17 ): # 각 블로그 주소 저장
        titles = driver.find_element_by_xpath('//*[@id="PostThumbnailAlbumViewArea"]/ul/li['+str(j)+']/a')

        title = titles.get_attribute('href')
        numbers = re.findall(r'\d+', title) #그냥 크롤링된 데이터를 바로 연결하면 mainframe을 찾기 어려움 -> 글 num을 추출해서 링크 조정
        numbers[0]
        blog_url='https://blog.naver.com/insoonlife/' + numbers[0]
        urls.append(blog_url)
        print(title)
print("url 수집 끝, 해당 url 데이터 크롤링")
```

```python
content_list = ''
title_list=''
for url in urls:  # 수집한 url 만큼 반복
    driver.get(url)  # 해당 url로 이동

    driver.switch_to.frame('mainFrame')
    overlays = ".se-fs-fs32.se-ff-system" # 내용 크롤링
    tit = driver.find_elements_by_css_selector(overlays)

    overlays = ".se-fs-fs16.se-ff-system"
    contents = driver.find_elements_by_css_selector(overlays)
    for x in tit:
        title =x.text
        content_list= content_list + title+':' # 블로그 글과 제목을 구분하기 위해 제목 크롤링 후에는 : 추가 -> 제목에서 일자 추출, 글에서 브랜드와 시간 추출
    for x in contents:
        content = x.text
        content_list = content_list +content +','
```

```python
import pandas as pd
import re
with open('C://Users//chlgo//cj//new_contents.txt','r',encoding='UTF-8') as f : # 파일 읽어오기 
    data = f.read().split(",")

data = pd.DataFrame(data)
```

```python
data
```

```python
title_date=[]
title_index = list(data[data[0].str.endswith(':')].index) # :로 끝나는 data를 title 인덱스 저장
event_date=[]
```

```python
for x in title_index: # title_index활용해서 title 추출
    event_string = data[0][x]
    title_date_list = re.findall('\d+',event_string) # 숫자만 추출 
    if len(title_date_list) ==1:
        title_date.append(title_date_list[0])
    else:
        title_date.append(title_date_list[0]+'/'+title_date_list[1]) # 월/ 일 형식으로 저장
```

```python
event_time=[]
event_brand=[]
```

```python
for x in range(len(title_index)-1): 
    event_data = data[0][title_index[x]:title_index[x+1]] # title_index를 기준으로 해당 날짜 event 나눠서 저장해 반복
    event_data = list(event_data)

    for index,value in enumerate(event_data): 

        if '오후' in value or '낮'in value or '오전' in value: # 시간을 나타내는 str이 존재 하면 시간으로 간주해 append
            event_time.append(value)
            try:

                brand_name = event_data[index+1] # 시간 바로 뒤에 존재하는 값이 brand  
                event_brand.append(brand_name) 
                event_date.append(title_date[x])
            except:
                pass
```

```python
print(len(event_date)) # 이벤트 날짜와
print(len(event_time)) # 이벤트 시간
print(len(event_brand)) # 이벤트 브랜드 간 len 비교로 데이터 확인

dict = {'event_date': event_date, 'event_time': event_time, 'event_brand': event_brand} # dict 형식으로 만들어 df으로 변환해 저장 
df = pd.DataFrame(dict)
```
